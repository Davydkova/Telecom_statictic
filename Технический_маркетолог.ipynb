### Какой коэффициент корреляции по отношению к цене больше: объема в литрах или мощности

```python
df_corr = df.corr()

ax = sns.heatmap(df_corr, annot=True)
ax.set_title('Матрица корреляции');
```

Коэффициент корреляции "объем, л" по отношению к выручки от продажи составляет **0.3** , а коэффициент "мощности, кВт" по отношению к выручки от продажи составляет **0.055**.

Коэффициент корреляции "объем, л" по отношению к стоимости 1 товара составляет **0.27** , а коэффициент "мощности, кВт" по отношению к стоимости 1 товара  составляет **0.41**.

**Большая корреляция по отношению к цене у объема**
